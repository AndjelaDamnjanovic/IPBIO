```python
import json
import sys
import re
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq

j=1
filenames = ["../trainSets/trainSet1.json", "../trainSets/trainSet2.json", "../trainSets/trainSet3.json", "../trainSets/trainSet4.json", "../trainSets/trainSet5.json"]

for filename in filenames:
    f=open(filename)
    podaci=json.load(f)
    f.close()

    file_out_id_alphafold_content = '../preprocessedFiles/idAlphafold_very_low_content' + str(j) +'.txt'
    
    j+=1

    n = len(podaci)

    print(n)

    alphafolds = []
    disprot_ids=[]

    for i in range(n):
        if (podaci[i].get('alphafold_very_low_content') is not None):
            alphafolds.append(podaci[i]['alphafold_very_low_content'])
            disprot_ids.append(podaci[i]['disprot_id'])

    #sada zelimo da napravimo .fasta fajl na osnovu podataka koje smo izvukli iz JSON fajla
#to radimo koriscenjem biblioteke Bio iz Pajtona

    with open(file_out_id_alphafold_content, 'w') as f_out:
        for i in range(len(alphafolds)):
            f_out.write(disprot_ids[i] + "\t" + str(alphafolds[i]) + "\n")
```

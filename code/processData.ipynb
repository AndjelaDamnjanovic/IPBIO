```python
import json
import sys
import re
import random
import numpy as np
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
```

```python
def processFile(podaci, n):
    #informacije o svakom proteinu pocinju poljem features, pa cemo po tome da delimo fajl
    #posto StatRepeats prima .fasta fajl, od datom JSON fajla zelimo da napravimo fasta fajl koji ce 
    #sadrzati id proteina i njegovu sekvencu
    names=[]
    sequences=[]
    disprot_ids=[]
    organisms=[]

    for i in range(n):
        names.append(podaci[i]['name'])
        sequences.append(podaci[i]['sequence'])
        disprot_ids.append(podaci[i]['disprot_id'])
        organisms.append(podaci[i]['organism'])

    #provera da li svi proteini imaju date osobine
    print(len(names))
    print(len(sequences))
    print(len(disprot_ids))
    print(len(organisms))
    return names, sequences, disprot_ids, organisms
```

```python
def shuffle(disprot_ids, seed):
    random.seed(seed)
    random.shuffle(disprot_ids)
    return disprot_ids
```

```python
def splitData(disprot_ids):
    trainSize = int(np.floor(60*len(disprot_ids)/100))
    print("Velicina trening skupa je: ", trainSize)

    train = disprot_ids[0:trainSize]

    test = disprot_ids[trainSize:]

    return train, test
```

```python
jsonTrain = []
jsonTest = []

testIds = []
trainIds = []
```

```python
trainFile = None
testFile= None
```

```python
def dumpToJson(testFile, trainFile):
    with open(testFile, "w") as f:
        f.write(json.dumps(jsonTest))
    with open(trainFile, "w") as f:
        f.write(json.dumps(jsonTrain))
```

```python
def openFile(filename, seed):
    print(filename)
    f=open(filename)
    podaci=json.load(f)
    f.close()

    n=podaci['size']
    print("Broj proteina koji se nalazi u fajlu ", filename, " je", n)
    #posto nam velicina fajla vise nije potrebna, vec su nam samo podaci potrebni, uklonicemo je
    podaci=podaci['data']
    names, sequences, disprot_ids, organisms = processFile(podaci, n)
    shuffle(disprot_ids, seed)

    train, test = splitData(disprot_ids)

    for i in range(n):
        if podaci[i]['disprot_id'] in test:# and podaci[i]['disprot_id'] not in testIds:
            jsonTest.append(podaci[i])
            testIds.append(podaci[i]['disprot_id'])
        elif podaci[i]['disprot_id'] in train:# and podaci[i]['disprot_id'] not in trainIds:
            jsonTrain.append(podaci[i])
            trainIds.append(podaci[i]['disprot_id'])
    #podaci
```
